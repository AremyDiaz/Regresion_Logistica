#Actividad de regresión logística

```python
#Cargamos librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as special
from scipy.optimize import curve_fit
import seaborn as sns
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#Se lee el archivo
df=pd.read_csv('BD_Socio formador (TrainingDataComplete).csv', index_col='Id')
df.head()
```

##Experimentos

Expermiento #1

```python
#verificar los valores sin repartirse de una columna
unico = np.unique(df['Age'])
unico
```

```python
df['Age'].median()
```

```python
df['Age_bool']=np.where(df['Age']>df['Age'].median(), 1, 0)
df
```

```python
#Declarar variables
Vars_Indep= df[['CURRENT_HOUSE_YRS']]
Var_Dep= df['Age_bool']

#Redefinir variables
x= Vars_Indep
y= Var_Dep

#Dividir el conjunto de datos en la parte de entrenamiento y prueba:
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state= None)

#Se escalan todos los datos
escalar = StandardScaler()

#Para realizar el escalamiento de las variables "x" tanto de entrenamiento como de prueba,
#solo se escala una variable y de tipo dicotonimca
x_train = escalar.fit_transform(x_train)
x_test = escalar.fit_transform(x_test)

#Definir el algoritmo a utilizar
from sklearn.linear_model import LogisticRegression
algoritmo = LogisticRegression()

#Entrenaar el modelo
algoritmo.fit(x_train, y_train)
```

```python
#Realizamos una predicción
y_pred = algoritmo.predict(x_test)
y_pred
```

```python
#Verifico la matriz de confusion
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusiòn: ')
print(matriz)
```

```python
#Calculo de la presición del modelo
#para tener una mejor precisión debemos de cambiar el numero de variables predictoras (y)
from sklearn.metrics import precision_score

precision = precision_score(y_test, y_pred, average= "binary", pos_label=0)
print('Precisión de modelo: ')
print(precision)
```

```python
#Calculo la exactitud del modelo
from sklearn.metrics import accuracy_score

exactitud = accuracy_score(y_test, y_pred)
print('Exactitud del modelo: ')
print(exactitud)
```

```python
#calculo de la sensibilidad del modelo
from sklearn.metrics import recall_score

sensibilidad=recall_score(y_test,y_pred, average='binary', pos_label=0)
print('sensibilidad del modelo:')
print(sensibilidad)
```

```python
#El puntaje F1 que es una combinación entre la precisión y la sensibilidad,
#Calculo el puntaje F1 del modelo
from sklearn.metrics import f1_score

puntajeF1=f1_score(y_test,y_pred,average='binary',pos_label=0)
print('Puntaje F1 del modelo')
print(puntajeF1)
```
